# Construindo gráficos a partir do banco de dados SQLite

Uma vez que os dados estejam armazenados, muitas operações podem ser realizadas. Estatísticas, gáficos, modelos. Agora vamos criar gráficos utilizando matplotlib, a principal biblioteca Python para confecção de gráficos. Para isso iremos utilizar o banco de dados criado nas etapas anteriores.

### Importando os pacotes, criando uma conexão, um cursor e uma função para leitura dos dados

```python
# Importando os pacotes

import sqlite3
import random
import datetime
import matplotlib.pyplot as plt

# essa linha de comando permite que o gráfico seja mostraado no mesmo jupyter notebook.

%matplotlib notebook 

# Criando uma conexão

conn = sqlite3.connect('dsa.db')

# Criando um cursor

c = conn.cursor()

# função para leitura de todos os dados

def leitura_todos_dados():
    c.execute('SELECT * FROM produtos')
    for linha in c.fetchall():
        print(linha)
        
leitura_todos_dados()
```

### Gerando uma função que constroi um gráfico de barras a partir dos dados da tabela produtos

```python
# Ddefinindo a função para gerar um gráfico de barras com os dados do banco de dados

def dados_grafico():
    c.execute('SELECT id, valor FROM produtos')
    ids = []
    valores = []
    dados = c.fetchall()
    for linha in dados:
        ids.append(linha[0])
        valores.append(linha[1])
    plt.bar(ids, valores)
    plt.show()
```

```python
# Gerando o gráfico

dados_grafico()
```
