# Fundamentos de Python - Bancos de Dados

### O que são bancos de dados?

Bancos de dados são conjuntos de dados organizados, de forma que se possa facilmente acessar, gerênciar e atualizar os dados. São coleção de dados interligados entre si e organizados para fornecer informações. Também pode ser definido como uma coleção de dados inter-relacionados, representando informações sobre um domíno específico. O banco de dados é uma forma de se armazenar dados para que estes gerem, posteriormente, informações.

### Dados x informações

Os dados são os fatos brutos, em sua forma primária. A informação consiste no agrupamento de dados de forma organizada para fazer sentido, para gerar conhecimento. 

### Sistema Gerenciado de Banco de Dados - SGBD

É o conjunto de programas de computador (softwares) responsáveis pelo gerenciamento de uma base de dados. Ele garante a consistência dos dados, controla o acesso, mantem os dados seguros, além de fornecer maios de acesso aos dados. Em banco de dados relacionais a interface de acesso ao SGBD é feita com comandos na linguagem SQL. O software ORACLE é um exemplo de SGBD. E por que usar um SGBD? Um SGBD oferece:
- Maior controle de redundância;
- Compartilhamento dados;
- Independência de dados;
- Maior segurança;
- Backup recuperação de falhas;
- Forçar restrições de integridade;
- Aumento da produtividade e disponibilidade;
- Flexibilidade e padronização.

### Bancos de dados relacionais

São o principal modelo para getão de bancos de dados pelas ultimas décadas. Eles representam e armazenam os dados em tabelas, organizadas em linhas e colunas. Eles são baseados em um ramo da teoria dos conjuntos algébricos, mais conhecido como álgebra relacional. Bancos de dados relacionais são gerenciados por um Relational Data Base Management System (RDBMS). Nesse sistema, todos os dados são armazenados em formato de tabelas com linhas e colunas e é a relação entre as tabelas que o torna um banco de dados relacional. 

A estrutura de um banco de dados relacional permite unir informações de diferentes tabelas atraves do uso de chaves estrangeiras, que são usadas para identificar exclusimente qualquer peça atômica de dados dentro dessas tabelas. As tabelas se conectam através de chaves estrangeiras, de modo a criar uma ligação entre os dados. Os bancos de dados relacionais são caracterizados por alguns conceitos fundamentais, tais como:
- Entidades (tabelas) e atributos;
- Registros (tuplas);
- Chaves (primária e estrangeira);
- Relacionamentos entre entidades (tabelas);
- Integridade Referencial;
- Normalização.

### Principais conceitos de bancos de dados relacionais
 
Em uma tabela, cada linha contem o mesmo conjunto de colunas. Em um banco de dados podem existir uma, ou centenas de tabelas. Cada linha formada por uma lista ordenada de colunas representa um __registro__, ou __tupla__, dentro da tabela. As colunas de uma tabela são chamadas de __atributos__. O conjunto de valores que um atributo pode assumir é chamado de __domínio__.

Um conceito importante em um banco de dados relacional é o conceito de __atributo chave__, que permite identificar e diferenciar uma tupla de outra. Através do uso de chaves é possível acelerar o acesso a elementos utilizando índices e com isso estabelecer o relacionamento entre as mútliplas tabelas de um sistema de banco de dados relacional. As tabelas relacionam-se umas as outras através de chaves. Uma chave é um conjunto de um ou mais atributos que determinam a unicidade de cada registro. 

__Chave primária (PK)__ - conluna(s) que garante(m) a unicidade das linhas. É a chave principal de uma tablea. Uma tabela só pode ter uma chave primária.

__Chave estrangeira (FK)__ - coluna(s) de uma tabela que contém valores da PK de outra tabela. Uma tabela pode ter mais de uma chave estrangeira.

As chaves garantem o relacionamento entre as tabelas. As tabelas se associam através de regras de relacionamento. 

__Índices__ - é uma lista ordenada de valores que apontam para os dados nas tabelas. Utilizado para agilizar a leitura de dados. Pode ser utilizado também para forçar a unicidade dos dados.

__Integridade referencial__ - é um conceito em que várias tabelas de banco de dados compartilham uma relação com base nos dados armazenados nas tabelas e essa relação deve ser coerente. Isso geralmente é imposto com ações de adição, exclusão e atualização em cascata. 

__Normalização de dados__ - é o processo de organização de campos e tabelas em um banco de dados relacional para minimizar a redundância e a dependência.

### Linguagem SQL

SQL significa Structure Query Language. é a linguagem utilizada para se comunicar com um banco de dados relacional. É usada para operações em banco de dados como consulta ao banco de dados, atualização, ou mesmo para remover dados. O servidor RDMS consiste em um banco de dados relacional, que dispõem de um conjunto de tabelas, que por sua vez contém dados com categorias e colunas pré-definidas. 

Os comandos da linguagem SQLse dividem em três grupos principais: __DDL__, __DML__, __DCL__. Os comandos do grupo DML (Data Manipulation Language) são utilizadas para o desenvolvimento. São elas:

- SELECT: pesquisa de dados;
- UPDATE: atualização de dados;
- DELETE: eliminação de dados;
- INSESRT: inserção de dados.

O segundo grupo de comandos, DDL (Data Definition Language) são utilizdas para suporte. São elas:

- CREATE: definição de um objeto (tabela, índice);
- ALTER: alteração de um objeto;
- DROP: eliminação de um objeto.

O terceiro grupo de comandos é o DCL (Data Control Language) que são utilizados tanto para suporte, como para desenvolvimento:

- GRANT: fornece um privilégio;
- REVOKE: tira um privilégio.

### Banco de dados SQLite

O SQLite é um engine (motor) de banco de dados SQL. É uma versão simplificada de banco de dados SQL. Um exemplo de versão completa de SQL é o MySQL. Todo o banco de dados SQLite é contido em um único arquivo. Ele carrega em memória apenas os arquivos que estão sendo usados. É mais simples de utilizar e mais leve. Ele pode sr usado quando recursos mais avançados de SGBD's não forem necessários, como por exemplo aplicações que executam em um smartphone. O SQLite é instalado junto com o anaconda. Ao contrário de outros bancos de dados, o SQLite nã requer um processo ou serviço sendo executado em um servidor. Todo acesso ao banco de dados é feito através de leitura/escrita aos arquivos do banco de dados, acesso este feito pelas bibliotecas do SQLite.

Qual é o uso apropriado do SQLite?

- Dipositivos embarcados (como em sistemas de internet das coisas);
- websites;
- análise de dados;
- cahce para dados de RDBMS;
- bancos de dados temporários e internos.

Quando não é apropriado usar o SQLite?

- Aplicações de cliente/servidor;
- websites com alto volume de tráfego;
- análise de datasets muito grandes;
- alta concorrência requerida;
- armazenar Big Data.

### Bancos de dados NoSQL

Bancos de dados relacionanais não foram porjetados para tratar grandes quantidades de dados (Big Data). Eles foram idealizados para tratar conjutos de dados que possam ser armazenados em linhas e colunas para que possam ser consultados pelo uso de queries utilizando a liguagem SQL. Os bancos de dados relacionais não possuem as funcionalidades necessárias para atender os requisitos de Big Data: dados gerados em grande volume, em alta velocida e em alta variedade. Essa launa é preenchicda por bancos de dados NoSQL, como por exemplo o MongoDB. 
Bancos de dados NoSQL são bancos de dados distribuídos e não relacionais que foram projetados para atender os requerimentos de Big Data. Eles oferecem uma arquitetura mais eficientes que facilitam a consulta de dados semi-estruturados e não estruturados. Os bancos de daos NoSQL oferecem 4 categorias principais de bancos de dados:

* Graph databases
* Document databases
* Key-values stores
* Column family stores

Consulte [aqui](https://nosql-database.org) uma lista completa de Bancos de Dados NoSQL.

### Banco de dados RDMS x NoSQL

Use RDBMS quando...                            | Use NoSQL quando...
:---                                           | :---
Aplicações centralizadas (ERP, CRM)            | Aplicações descentralizadas (Web, Mobile, Big Data, IoT)
Alta disponibilidade for necessária            | Disponibilidade contínua, sem interrupção
Dados gerados em velociadade moderada          | Dados gerados em alta velocidade (sensores)
Dados gerados a partir de poucas fontes        | Dados gerados a partir de múltiplas fontes
Dados estruturados                             | Dados semi ou não-estruturados
Transações complexas                           | Transações simples
Volume moderado de dados                       | Alto volume de dados


### Banco de dados MongoDB

É um banco de dados orientado a documento programado em linguagem C++. Ele substitui o conceito de 'linha' como em banco de dados relacionais por um modelo mais flexível, o 'documento'. O MongoDB é openSource e um dos lideres no segmento de banco de dados NoSQL. Algumas das principais características do MongoDB são:

- Indexação: suporta índices secundários e permite a construção de queries mais velozes;
- Agragação: permite a construção de agragações complexas de dados, otimizando assim o desempenho;
- Tipos de dados especiais: suporta coleções do tipo time to live para dados que expiram em um determinado tempo;
- Armazenamento: suporta armazenamento de rgandes quantidades de dados

Onde usar o MongoDB

- Big data;
- Gestão de conteúdo;
- Infraestrutura social e Mobile;
- Gestão de dados de usuários;
- Data Hub;
- Capturar Stream de dados para análise
